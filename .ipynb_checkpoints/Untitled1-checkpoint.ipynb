### カンスーたち

```python
def make_input_x(n, p):
    global Ntrue, Nfalse,TRUE, FALSE, input_x
    
    degree =360*p/2
    Ntrue, Nfalse = int(n*p), int(n-n*p)
    TRUE, FALSE = np.zeros([Ntrue, 2]), np.zeros([Nfalse, 2])
    
    true= np.random.uniform(-degree/180*np.pi, degree/180*np.pi, Ntrue)
    false = np.random.uniform(degree/180*np.pi, (360-degree)/180*np.pi, Nfalse)
    
    TRUE[:, 0], TRUE[:, 1] = np.cos(true), np.sin(true)
    FALSE[:, 0], FALSE[:, 1] = np.cos(false), np.sin(false) 
    
    input_x = np.r_[TRUE, FALSE]

def make_input_y(TRUE, FALSE):
    global input_y
    a = np.array([1, 2, 0]) 
    fx_true = a[0] + a[1]*TRUE[:, 0] 
    fx_false = a[0] + a[1]*FALSE[:, 0] +np.random.uniform(-10,10, Nfalse)
    
    input_y = np.r_[fx_true, fx_false]
    
    
def make_data():
    global DATA
    DATA = np.zeros([n, 6])
    make_input_x(n, p)
    make_input_y(TRUE, FALSE)
    data = np.c_[input_x, input_y]
    DATA[:, :3] = data
```
