```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("darkgrid")
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=Axes3D(fig)

ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("y")

N=120

s=np.random.uniform(0,2*np.pi,N)
x1=np.cos(s)
x2=np.sin(s)

X1=x1.reshape((N,1))
X2=x2.reshape((N,1))

x=np.c_[x1,x2]

theta0=np.random.normal(1,1)
theta1=np.random.normal(2,1)
theta2=np.random.normal(0,1)
phat=0.3


def f(i):
    return x[i,0]+2*x[i,1]

def Y1(i):
    if x[i,0]>0.5:
        y=np.array([f(i)])
        ax.scatter(x[i,0],x[i,1],y,edgecolors='k',lw=1,color="b")
        return y
    else:
        y=np.array([f(i)+np.random.uniform(-10,10)])
        ax.scatter(x[i,0],x[i,1],y,lw=1,color="b",marker='x')
        return y

Y=np.array([Y1(i) for i in range(N)])

def P(i):
    return theta0+theta1*X1[i]+theta2*X2[i]

Yhat=np.array([P(i) for i in range(N)])

error=np.abs(Yhat-Y)
print(error)
sort_error=np.sort(error,axis=0)
print(sort_error)

plt.show()
```

```python
import numpy as np
import pandas as pd
import sklearn
import sklearn.svm as svm
import matplotlib.pyplot as plt
%matplotlib inline

N=120

s=np.random.uniform(0,2*np.pi,N)
x1=np.cos(s)
x2=np.sin(s)

X1=x1.reshape((N,1))
X2=x2.reshape((N,1))

x=np.c_[x1,x2]

theta0=np.random.normal(1,1)
theta1=np.random.normal(2,1)
theta2=np.random.normal(0,1)
phat=0.3


def f(i):
    return x[i,0]+2*x[i,1]

def Y1(i):
    if x[i,0]>0.5:
        y=np.array([f(i)])
        return y
    else:
        y=np.array([f(i)+np.random.uniform(-10,10)])
        return y

Y=np.array([Y1(i) for i in range(N)])

def P(i):
    return theta0+theta1*X1[i]+theta2*X2[i]

Yhat=np.array([P(i) for i in range(N)])

err=Yhat-Y
r=np.abs(err[range(N),0])
er=r.reshape(N,1)
errorx=np.c_[er,x]

sort_error=errorx[errorx[:,0].argsort(),:]
n=int(N*phat)
x1=np.hsplit(sort_error,[1,3])

X=x1[1]
y=np.array([1]*n+[-1]*(N-n))

print(X)
print(y)
est=svm.LinearSVC()
est.fit(X,y)

xx,yy=np.meshgrid(np.linspace(-2,2,1000),
                  np.linspace(-2,2,1000))

def plot_predict(est,title):
    Z=est.predict(np.c_[xx.ravel(),
                        yy.ravel()])
    Z=Z.reshape(xx.shape)
    cmap=plt.cm.Blues
    
    fig,ax=plt.subplots(1,1,figsize=(5,5))
   
    ax.contourf(xx,yy,Z,
               cmap=plt.cm.terrain,
               alpha=0.5)
    ax.scatter(X[:,0],X[:,1],
               s=50,c=.5+.5*y,
               edgecolors='k',
               lw=1,cmap=cmap,
               vmin=0,vmax=1)
    ax.axhline(0,color='k',ls='--')
    ax.axvline(0,color='k',ls='--')
    ax.axis([-2,2,-2,2])
    ax.set_title(title)

ax=plot_predict(est,"linear SVC")


```

```python
import numpy as np
import pandas as pd
import sklearn
import sklearn.svm as svm
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline

N=120

s=np.random.uniform(0,2*np.pi,N)
x1=np.cos(s)
x2=np.sin(s)

X1=x1.reshape((N,1))
X2=x2.reshape((N,1))

x=np.c_[x1,x2]

theta0=np.random.normal(1,1)
theta1=np.random.normal(2,1)
theta2=np.random.normal(0,1)
thetao=np.c_[theta0,theta1,theta2].reshape(1,3)
phat=0.3
print(thetao)

k=70
for l in range(k):
    
    def f(i):
        return x[i,0]+2*x[i,1]

    def Y1(i):
        if x[i,0]>0.5:
            y=np.array([f(i)])
            return y
        else:
            y=np.array([f(i)+np.random.uniform(-10,10)])
            return y

    Y=np.array([Y1(i) for i in range(N)])

    def P(i):
        return thetao[0,0]+thetao[0,1]*X1[i]+thetao[0,2]*X2[i]

    Yhat=np.array([P(i) for i in range(N)])

    err=Yhat-Y
    r=np.abs(err[range(N),0])
    er=r.reshape(N,1)
    errorxY=np.c_[er,x,Y]

    sort_error=errorxY[errorxY[:,0].argsort(),:]
    n=int(N*phat)
    x1=np.hsplit(sort_error,[1,3])
    X=x1[1]
    y=np.array([1]*n+[-1]*(N-n))

    est=svm.LinearSVC()
    est.fit(X,y)

    y1=est.predict(X)
    Y1=y1.reshape(N,1)

    xy=np.hsplit(sort_error,[1,4])
    xY=xy[1]
    Y1s=np.c_[Y1,xY]
    y1s=Y1s[Y1s[:,0].argsort(),:]

    j=np.count_nonzero(y1s[:,0]==-1)

    y1S=np.vsplit(y1s,[j,N])
    Y1S=y1S[1]
    YY=np.hsplit(Y1S,[1,3])
    xx=YY[1]
    yy=YY[2]

    clf=linear_model.LinearRegression()
    clf.fit(xx,yy)
    b=clf.coef_.reshape(1,2)
    a=clf.intercept_
    thetan=np.c_[a,b].reshape(1,3)
    
    if  np.linalg.norm((thetan-thetao)/thetao)<=0.01:
        print(l)
        break
    else:
        thetao=thetan
        print(thetao)
```
